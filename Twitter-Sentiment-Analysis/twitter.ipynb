```python
import json
import re
import pandas as pd
```

```python
with open('new.json','r') as filename:
    data=json.load(filename)

df=pd.DataFrame(data)
```

```python
hosttweets=df[df['text'].str.contains('host')]
```

```python
bestTweets=df[df['text'].str.contains('[B|b]est .* [in|-] [\w]',regex=True)]
```

```python
host_r = '((?:[A-Z][a-z]+)(?: [A-Z][a-z]+)+)'
found_host = {}
found_host_set=set()
# print(hosttweets)
for i in hosttweets['text'] :
    
    h=re.findall(host_r,i)
    
    for i in h:
        if i not in found_host_set:
            found_host_set.add(i)
            found_host[i]=1
        else :
            found_host[i]=found_host[i]+1


found_host=dict(reversed(sorted(found_host.items(),key=lambda item :item[1])))
found_host
```

```python
final_hosts = []
count = 0
for i in found_host:
    if count > 1:
        break
    count += 1
    final_hosts.append(i)

final_hosts
```

```python
presentTweets = df[df['text'].str.contains('present')]
```

```python
regx = "(?:(?<=presenter)(?:.*)(?: [A-Z][a-z]+)+(?: [A-Z][a-z]+)+ and (?: [A-Z][a-z]+)+(?: [A-Z][a-z]+))|(?:(?: [A-Z][a-z]+)+(?: [A-Z][a-z]+)+ (?=are presenting))|(?:(?<=presented by)(?: [A-Z][a-z]+)+(?: [A-Z][a-z]+)+ )|(?:(?: [A-Z][a-z]+)+(?: [A-Z][a-z]+)+ (?=present))|(?:(?: [A-Z][a-z]+)+(?: [A-Z][a-z]+)+ (?=will present))"
prsnt_set=set()
prsnt_dict={}
for j in presentTweets['text']:
    lst=re.findall(regx,j)
    for i in lst:
        if i not in prsnt_set:
            prsnt_set.add(i)
            prsnt_dict[i]=1
        else:
            prsnt_dict[i] = prsnt_dict[i] + 1
# print(prsnt_dict)

final_pres=sorted(prsnt_dict.items(), key = lambda x:(x[1]), reverse=True)
final_pres
```
